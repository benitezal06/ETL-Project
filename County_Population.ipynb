```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile
import numpy as np
from scipy.stats import sem
import os
```

```python
#Bring in County Population Data
county_pop_df = pd.read_excel('Data\population_by_jurisdiction_and_by_county_-_1970_to_2018_-_09-07-2018.xlsx', sheet_name='Population by County')
county_pop_df
```

```python
#Bring in CoC Data
coc_df=pd.read_csv('Data\coc_county_xref.csv', header=None)
coc_df
```

```python
# Rename County and CoC Number Column Headers
coc_df.rename(columns = {0:'COC Number'}, inplace = True) 
coc_df.rename(columns = {1:'County'}, inplace = True) 
coc_df
```

```python
# Only show records for 2018
is_2018 =  county_pop_df['Year']==2018
print(is_2018.head())
pop_2018 = county_pop_df[is_2018]
print(pop_2018)
```

```python
pop_2018.count()
```

```python
#COC Data marge with County Population Data
COC_merge=pd.merge(coc_df, pop_2018, how='left') 

#Drop Year Column - Not Needed Anymore
del COC_merge['Year']
COC_merge
```

```python
pop_2018_COC = COC_merge.groupby(['COC Number'], ).sum()
pop_2018_COC
```

```python
pop_2018_COC.to_csv('Data\COC_Population_2018.csv')
```
